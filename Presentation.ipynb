# Optimization of the queries

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from matplotlib import rcParams
%matplotlib inline
rcParams['figure.figsize'] = 25,12
```

## Query 1 

### improvement: 
    
    - 8.4 sec --> 4.9 sec
    - 7.2 sec --> 0.8 sec

_athlete with most first places_

```python
img_A = mpimg.imread('1base.png'); img_B = mpimg.imread('1optim.png')
ax1 = plt.subplot(1,2,1); ax1.set_title('Non optimized', size = 20); ax1.imshow(img_A); plt.sca(ax1)
ax2 = plt.subplot(1,2,2); ax2.set_title('Optimized', size = 20); ax2.imshow(img_B); plt.sca(ax2)
```
