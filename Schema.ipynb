```python
import pandas as pd
import numpy as np
from datetime import datetime
from tqdm import tqdm
from collections import defaultdict
from pymongo import MongoClient
import json
import pickle

def save_pickle(element, path):
    with open(f"{path}", 'wb') as f:
        pickle.dump(element, f, pickle.HIGHEST_PROTOCOL)

def load_pickle(path):
    with open(f"{path}", 'rb',) as f:
        return pickle.load(f)
```

```python
df = pd.read_csv(r'/home/arturo/Desktop/DM/HW2/csv/results.csv', low_memory = False)
df.fillna('NA', inplace=True)
```

```python
docs = []
for i in tqdm(range(len(df))):
   doc = {}
   doc['_id'] = df.loc[i, '_id']
   doc['athlete'] = {'Name': df.loc[i, 'Name'], 'Sex': df.loc[i, 'Sex'], 'Age': df.loc[i, 'Age'], 'BodyweightKg': df.loc[i, 'BodyweightKg']}
   doc['meet'] = {'Date': df.loc[i, 'Date'], 'MeetCountry': df.loc[i, 'MeetCountry'], 'MeetName': df.loc[i, 'MeetName']}
   doc['competition'] = {'Event': df.loc[i, 'Event'], 'Division': df.loc[i, 'Division']}
   doc['squat'] = {'Squat1Kg': df.loc[i, 'Squat1Kg'], 'Squat2Kg': df.loc[i, 'Squat2Kg'], 'Squat3Kg': df.loc[i, 'Squat3Kg'], 'Squat4Kg': df.loc[i, 'Squat4Kg']}
   doc['bench'] = {'Bench1Kg': df.loc[i, 'Bench1Kg'], 'Bench2Kg': df.loc[i, 'Bench2Kg'], 'Bench3Kg': df.loc[i, 'Bench3Kg'], 'Bench4Kg': df.loc[i, 'Bench4Kg']}
   doc['deadlift'] = {'Deadlift1Kg': df.loc[i, 'Deadlift1Kg'], 'Deadlift2Kg': df.loc[i, 'Deadlift2Kg'], 'Deadlift3Kg': df.loc[i, 'Deadlift3Kg'], 'Deadlift4Kg': df.loc[i, 'Deadlift4Kg']}
   doc['scores'] = {'Wilks': df.loc[i, 'Wilks'], 'McCulloch': df.loc[i, 'McCulloch'], 'Glossbrenner': df.loc[i, 'Glossbrenner'], 'IPFPoints': df.loc[i, 'IPFPoints']}
   doc['results'] = {'TotalKg': df.loc[i, 'TotalKg'], 'Place': df.loc[i, 'Place'], 'Equipment': df.loc[i, 'Equipment']}
   docs.append(doc)
```

```python
docs = load_pickle('/home/arturo/Desktop/DM/HW2/docs.pickle')
```

```python
client = MongoClient("mongodb://localhost:27017")
db = client.powerlifting
coll = db['records']

for doc in tqdm(docs):
    coll.insert_one(doc)
```
